```python
# Juntamos los data frames train y train2 a través de su ID

train_df = pd.merge(train, train2, on='Employee_ID', how='left')
print(train_df.shape)
train_df.head()
```

```python
#aqui juntamos los test

test_df = pd.merge(test, test2, on='Employee_ID', how='left')
print(test_df.shape)
test_df.head()
```

```python
#Aqui concatenamos train total y test total
df = pd.concat([train_df, test_df], axis=0, ignore_index=True)
print(df.shape)
df.head()
```

Para todos los registros que no tienen un valor para un sistema operativo, ya sea de su computadora o de su dispositivo móvil, asígnales el valor “no registrado“.


```python
print(df[' Computer_OS'].value_counts())
print(df[' Mobile_OS'].value_counts())
```

```python
print(df[' Computer_OS'].isna().sum())
print(df[' Mobile_OS'].isna().sum())
```

```python
df[' Computer_OS'].fillna('No registrado', inplace=True)
df[' Mobile_OS'].fillna('No registrado', inplace=True)
```

```python
print(df[' Computer_OS'].value_counts())
print(df[' Mobile_OS'].value_counts())
```

Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:
El porcentaje de personas que usan Windows, Mac, Linux o "No registrado" como su sistema operativo en su computadora. Se esperan 4 renglones en esta respuesta.

```python
compu_df = df.groupby(by=[' Computer_OS']).agg({'Employee_ID':np.size}).rename(columns={"Employee_ID" : "cuantos"}).reset_index()
compu_df
```

```python
compu_df['porcentaje'] = compu_df['cuantos'] / compu_df['cuantos'].sum() * 100
compu_df
```

El porcentaje de personas que usan iOS, Android o "No registrado" como su sistema operativo en su dispositivo móvil. Se esperan 3 renglones en esta respuesta.

```python
cel_df = df.groupby(by=[' Mobile_OS']).agg({'Employee_ID':np.size}).rename(columns={"Employee_ID" : "cuantos"}).reset_index()
cel_df
```

```python
cel_df['porcentaje'] = cel_df['cuantos'] / cel_df['cuantos'].sum() * 100
cel_df
```

La edad promedio de la gente que utiliza cada sistema operativo de escritorio.

```python
edad_df = df.groupby(by=[' Computer_OS']).agg({'Age': np.sum, 'Employee_ID':np.size}).rename(columns={"Age":"Suma Edad", "Employee_ID" : "cuantos"}).reset_index()
edad_df
```

```python
edad_df['promedio edad'] = edad_df['Suma Edad'] / edad_df['cuantos']
edad_df
```

El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio.

```python
educacion_df = df.groupby(by=[' Computer_OS']).agg({'Education_Level': np.sum, 'Employee_ID':np.size}).rename(columns={"Education_Level":"Suma Educación", "Employee_ID" : "cuantos"}).reset_index()
educacion_df
```

```python
educacion_df['promedio educacion'] = educacion_df['Suma Educación'] / educacion_df['cuantos']
educacion_df
```

El número total de tickets generados por la gente, con cada sistema operativo de escritorio.

```python
df.columns
```

```python
total_tickets_df = df.groupby(by=[' Computer_OS']).agg({' Computer_tickets': np.sum, 'Employee_ID':np.size}).rename(columns={" Computer_tickets":"Suma Tickets", "Employee_ID" : "cuantos"}).reset_index()
total_tickets_df
```

```python
# esto no lo pide pero lo agrego
total_tickets_df['promedio tickets'] = total_tickets_df['Suma Tickets'] / total_tickets_df['cuantos']
total_tickets_df
```
